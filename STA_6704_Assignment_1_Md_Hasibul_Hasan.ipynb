# Question 2

```python
#Define f(x)
def f(x):
    return -2*np.exp((-(x[0]-1)**2 - (x[1]-1)**2)) - np.exp(-(x[0]+2)**2 - (x[1]+3)**2)
#Differntiate with respect to x1
def dfdx(x):
    return 4*(x[0]-1)*np.exp((-(x[0]-1)**2 - (x[1]-1)**2)) + 2*(x[0]+2)*np.exp((-(x[0]+2)**2 - (x[1]+3)**2))
#Differentiate with respect to x2
def dfdy(x):
    return 4*(x[1]-1)*np.exp((-(x[0]-1)**2 - (x[1]-1)**2)) + 2*(x[1]+3)*np.exp((-(x[0]+2)**2 - (x[1]+3)**2))
#Get gradient array
def gradf(x):
    return (np.array([dfdx(x), dfdy(x)]))
```

# Question 3

```python
#Define x1,x2 and f_x to append later
x1_1_1 = []
x2_1_1 = []
fx_1_1 = []
x1_1_2 = []
x2_1_2 = []
fx_1_2 = []
x1_1_3 = []
x2_1_3 = []
fx_1_3 = []
x1_2_1 = []
x2_2_1 = []
fx_2_1 = []
x1_2_2 = []
x2_2_2 = []
fx_2_2 = []
x1_2_3 = []
x2_2_3 = []
fx_2_3 = []
```

```python
#For init_value = [0,0]
def grad_descent(f, gradf, init_value, alpha):
    Precision = 1e-6
    prev_value = init_value-10*Precision #Keep the initial  previous value 10 fold precison smaller than initial value
    value = init_value.copy()
    max_iteration = 10000
    iteration = 0
    while norm(value - prev_value) > Precision and iteration < max_iteration:
        prev_value = value.copy()
        value -= alpha*gradf([value[0], value[1]])
        print ('Iteration= ',iteration,'[x1,x2]= ',value,'f([x1 x2])= ', f([value[0], value[1]]))
        if alpha == 0.1:
            x1_1_1.append(value[0])
            x2_1_1.append(value[1])
            fx_1_1.append(f([value[0], value[1]]))
        elif alpha == 0.01:
            x1_1_2.append(value[0])
            x2_1_2.append(value[1])
            fx_1_2.append(f([value[0], value[1]]))
        elif alpha == 0.001:
            x1_1_3.append(value[0])
            x2_1_3.append(value[1])
            fx_1_3.append(f([value[0], value[1]]))
            
        iteration += 1
    
    return value
```

```python
#For init_value = [-1,-1]
def grad_descent2(f, gradf, init_value, alpha):
    Precision = 1e-6
    prev_value = init_value-10*Precision #Keep the initial  previous value 10 fold precison smaller than initial value
    value = init_value.copy()
    max_iteration = 100000
    iteration = 0
    while norm(value - prev_value) > Precision and iteration < max_iteration:
        prev_value = value.copy()
        value -= alpha*gradf([value[0], value[1]])
        print ('Iteration= ',iteration,'[x1,x2]= ',value,'f([x1 x2])= ', f([value[0], value[1]]))
        if  alpha == 0.1:
            x1_2_1.append(value[0])
            x2_2_1.append(value[1])
            fx_2_1.append(f([value[0], value[1]]))
        elif alpha == 0.01:
            x1_2_2.append(value[0])
            x2_2_2.append(value[1])
            fx_2_2.append(f([value[0], value[1]]))
        elif alpha == 0.001:
            x1_2_3.append(value[0])
            x2_2_3.append(value[1])
            fx_2_3.append(f([value[0], value[1]]))
            
        iteration += 1
    
    return value
```
